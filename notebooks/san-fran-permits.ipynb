**TIDF Compliance** is the known as the permit required to pay the Transit Impact Development Fee.
**Voluntary Soft-Story Retrofit** means the property owner has chosen to retrofit the building to increase its earthquake resistance.
**The Unit Suffix** refers to an additional identifier in a building's address.
Those columns have many missing values because these factors are not applicable for most buildings. Moreover, inconsistent data entry could contribute to these missing values.

Dropping the missing values isn't a good option, especially when there are multiple rows that have over 99% missing values. However, it is reasonable to drop such columns.
An alternative to dropping the columns is to impute or fill missing data. For example, with `Unit Suffix` column, we can fill missing data with either `None` or `Single Unit`

### Handling Missing Values:

**Aproach**
For columns with data type as "object", they usually represent categorical or text data.  
I will fill these missing values with "Unknown".  
For columns with numeric data, I will handle the missing values with two methods: back fill + forward fill and mean fill.


```python
df_bffill = df.copy()
df_meanfill = df.copy()

object_columns = df.select_dtypes(include=['object']).columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns

# Fill missing string values with 'Unknown'
df_bffill[object_columns] = df_bffill[object_columns].fillna('Unknown')
df_meanfill[object_columns] = df_meanfill[object_columns].fillna('Unknown')

# Fill missing numerical values with the mean of the column
df_meanfill[numerical_columns] = df_meanfill[numerical_columns].fillna(df_meanfill[numerical_columns].mean().round(0))

# Fill missing numerical values with the bffill method
df_bffill[numerical_columns] = df_bffill[numerical_columns].bfill().ffill()
```

Double check to see if there are any missing values in the imputed DataFrame

```python
print("\nMissing Values After Fill:")
print("Mean Fill:")
print(df_meanfill.isna().sum().sum())
print("Bffill Fill:")
print(df_bffill.isna().sum().sum())
```

### Statistical Impact of Filling Data.

```python
# Statistical impact of filling missing values
# Mean Fill
mean_difference = df_meanfill[numerical_columns].mean() - df[numerical_columns].mean()
print("\nMean Difference:")
print(mean_difference.round(3))

std_difference = df_meanfill[numerical_columns].std() - df[numerical_columns].std()
print("\nStandard Deviation Difference:")
print(std_difference.round(3))

corr_difference = df_meanfill[numerical_columns].corr() - df[numerical_columns].corr()
print("\nCorrelation Difference:")
print(corr_difference.round(3))
```

```python
# Bffill Fill
mean_difference = df_bffill[numerical_columns].mean() - df[numerical_columns].mean()
print("\nMean Difference:")
print(mean_difference.round(3))

std_difference = df_bffill[numerical_columns].std() - df[numerical_columns].std()
print("\nStandard Deviation Difference:")
print(std_difference.round(3))

corr_difference = df_bffill[numerical_columns].corr() - df[numerical_columns].corr()
print("\nCorrelation Difference:")
print(corr_difference.round(3))
```
